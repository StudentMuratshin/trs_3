```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

#     
t = np.linspace(0, 1, 100)
x = np.linspace(0, 2, 100)
t, x = np.meshgrid(t, x)
y = np.sin(np.pi * x * 0.5) * (np.sin(np.pi * t)) ** 2

#  
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(t, x, y)

#    
ax.set_xlabel('y')
ax.set_ylabel('x')
ax.set_zlabel('u')
#plt.title('(1-np.exp(-t))*np.sin(2*np.pi*x)+x**2')

#  
plt.show()
```

```python
import numpy as np

#    
data = np.genfromtxt('u1_out.csv', delimiter=' ')

#     
time_steps = len(np.unique(data[:, 0]))
space_steps = len(np.unique(data[:, 1]))
solution = data[:, 2].reshape(time_steps, space_steps)

#    
x = np.linspace(0, 1, space_steps)
y = np.linspace(0, 1, time_steps)
X, Y = np.meshgrid(x, y)

#  3D 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(X, Y, solution)
ax.set_xlabel('Space')
ax.set_ylabel('Time')
ax.set_zlabel('Solution')
plt.show()
```
